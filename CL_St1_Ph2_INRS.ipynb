## Importing the debates into a dataframe

```python
# Initialising the dataframe
df_debates = pd.DataFrame()

# Concatenating the dataframes
for debate in debates_list:
    debate_filename = f'{input_directory}/{debate}.jsonl'
    df = pd.read_json(debate_filename, lines=True)
    df_debates = pd.concat([df_debates, df], ignore_index=True)
```

```python
# Combine all text data into a single string
all_text = ' '.join(df_debates['Text'])

# Split the combined text into individual words and convert to lower case
words = all_text.lower().split()

# Remove punctuation and other non-alphabetic characters
#words = [re.sub(r'\W+', '', word) for word in words]

# Convert the list of words into a set to get unique words
unique_words = set(words)

# Convert the set back to a sorted list (optional)
unique_words = sorted(unique_words)

with open('unique_words.txt', 'w', encoding='utf8', newline='\n') as file:
    for word in unique_words:
        file.write(f'{word}\n')
```

```python
# List of patterns to remove
patterns_to_remove = [
    r'\[applause\]',
    r'\[cross talk\]',
    r'\[crosstalk\]',
    r'\[laughing\]',
    r'\[Laughter and applause\]',
    r'\[laughter\.\]',
    r'\[laughter\]',
    r'\[sic\]',
    r'\[to Biden\]',
    r'\[twenty-seven-minute delay\]',
    r'\[unintelligible\]'
]

# Defining a function to remove specific patterns
def remove_specific_patterns(input_line):
    for pattern in patterns_to_remove:
        input_line = re.sub(pattern, '', input_line, flags=re.IGNORECASE)
    return input_line

# Removing specific patterns from the 'text' column
df_debates['Text'] = df_debates['Text'].apply(remove_specific_patterns)
```

```python
# Assuming df_debates is your DataFrame
new_row = {
    'Title': 'New Debate Title',
    'Debate': 'New Debate',
    'Date': '2024-07-17',
    'Participants': 'Participant1, Participant2',
    'Moderators': 'Moderator1',
    'Speaker': 'Speaker1',
    'Text': 'This is the text of the new debate.'
}

# Convert the new row to a DataFrame
new_row_df = pd.DataFrame([new_row])

# Use concat to append the new row
df_debates = pd.concat([df_debates, new_row_df], ignore_index=True)
df_debates
```
